### Data cleansing

```python
df['label']=np.where(df['diagnosis']=='M',1,0)
df.drop(['Unnamed: 32','id','diagnosis'],axis=1,inplace=True)
df.info()
```

```python
#heatmap of correlation between features
import seaborn as sns
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(25,10))
corr=df.corr()
sns.heatmap(corr,annot=True,cmap="Blues")
plt.show()
```

```python
sns.countplot(x=df['label'])
plt.title('Diagnosis')
plt.show()
```

### Data processing

```python
#define x and y
x=df.iloc[:,:30]
y=df.iloc[:,30]
print(x.shape,y.shape)
```

```python
# data standardization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=scaler.fit_transform(x)
```

### K-Nearest Neighbors

```python
#find the best hyperparameter,determine the n_neighbors
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
param_grid={'n_neighbors':np.arange(1,50)}
knn=KNeighborsClassifier()
knn_cv=GridSearchCV(knn,param_grid,cv=5)
knn_cv.fit(x,y)
print(knn_cv.best_params_)
print(knn_cv.best_score_)
```

```python
score_table=pd.DataFrame(columns=["Model","Accuracy"])
```

```python
#train test split, build and train model
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=42)
knn=KNeighborsClassifier(n_neighbors=9)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
score=knn.score(x_test,y_test)
print("KNN test score:", score)

new_row={"Model": "KNN", "Accuracy": score}
score_table=score_table.append(new_row,ignore_index=True)
```

```python
#cross validation - accuracy
from sklearn.model_selection import cross_val_score
cv_results=cross_val_score(knn,x,y,cv=5)
print("The average cv score is:{}".format(np.mean(cv_results)))
```

```python
#cross validation - recall
from sklearn.model_selection import cross_val_score
cv_results=cross_val_score(knn,x,y,cv=5,scoring='recall')
print("The average cv recall score is:{}".format(np.mean(cv_results)))
```

```python
#classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import accuracy_score, roc_curve, roc_auc_score
def roc_auc_curve(model,model_name):
    pred_prob=model.predict_proba(x_test)
    fpr,tpr,thresh=roc_curve(y_test,pred_prob[:,1],pos_label=1)
    score=roc_auc_score(y_test,pred_prob[:,1])
    
    plt.figure(figsize=(10,7))
    plt.plot(fpr,tpr,linestyle="--",color="blue",label="ROC curve(area=%1f)"%score)
    plt.title(f"{model_name} ROC curve")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.legend()
    plt.show()
```

```python
roc_auc_curve(knn,"K Nearest Neighbors")
```

### Principal Component Analysis (PCA)

```python
from sklearn.decomposition import PCA
pca=PCA()
pca.fit(x)

plt.figure(figsize=(10,5))
plt.plot(pca.explained_variance_ratio_)
plt.title('N components and explained variance ratio')
plt.xlabel('N components')
plt.ylabel('Expained variance ratio')
plt.show()
```

It looks like the first 5 components explained most of the information, it actually explained 99.9% of the variance. 

```python
pca=PCA(n_components=5)
x=pca.fit_transform(x)
pca.fit(x)
print(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_.sum())
```

### Logistic Regression

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#find the best hyperparameter
from sklearn.linear_model import LogisticRegression

param_grid={'C':[0.0001,0.001,0.01,0.1,1,10]}
lr=LogisticRegression()
lr_cv=GridSearchCV(lr,param_grid,cv=5)
lr_cv.fit(x,y)
print(lr_cv.best_params_)
print(lr_cv.best_score_)
```

```python
#build logistic regression model
lr=LogisticRegression(C=10)
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
score=lr.score(x_test,y_test)
print("Logistic regression test score:", score)

new_row={"Model": "Logistic Regression", "Accuracy": score}
score_table=score_table.append(new_row,ignore_index=True)
```

```python
#classification report
print(classification_report(y_test,y_pred))
```

```python
roc_auc_curve(lr,"Logistic Regression")
```

### Random Forest

```python
#find the best hyperparameter
from sklearn.ensemble import RandomForestClassifier

param_grid={"n_estimators":[100,200,300,400,500]}
rfc=RandomForestClassifier()
rfc_cv=GridSearchCV(rfc,param_grid,cv=5)
rfc_cv.fit(x,y)
print(rfc_cv.best_params_)
print(rfc_cv.best_score_)
```

```python
#build random forest classifier model
rfc=RandomForestClassifier(n_estimators=200)
rfc.fit(x_train,y_train)
y_pred=rfc.predict(x_test)
score=rfc.score(x_test,y_test)
print("Random forest classifier test score:", score)

new_row={"Model": "Random forest classifier", "Accuracy": score}
score_table=score_table.append(new_row,ignore_index=True)
```

```python
#classification report
print(classification_report(y_test,y_pred))
```

```python
roc_auc_curve(rfc,"Random Forest Classifer")
```

```python
score_table.sort_values(by=['Accuracy'],ascending=False)
```

Conclusion: Logistic Regression has the best accuracy.
